```python
import matplotlib.pyplot as plt

# Group the data by review count and calculate the average rating for each group
average_rating = yelp_foursquare_df.groupby("review_count")["rating"].mean()

# Create a line plot
plt.figure(figsize=(8, 6))
plt.plot(average_rating.index, average_rating.values)
plt.title("Average Rating by Review Count")
plt.xlabel("Review Count")
plt.ylabel("Average Rating")
plt.show()
```
