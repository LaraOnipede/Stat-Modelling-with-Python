# Database

Put all your results in an SQLite3 database (remember, SQLite stores its databases as files in your local machine - make sure to create your database in your project's data/ directory!)

```python
import sqlite3
from sqlite3 import Error
```

```python
def create_connection(path):
    connection = None
    try:
        connection = sqlite3.connect(path)
        print("connection to sqlite DB sucessful")
    except Error as e:
        print(f"The error '{e}' occurred ")
        
    return connection
```

```python
def execute_query(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        connection.commit()
        print("Query executed successfully")
    except Error as e:
        print(f"The error '{e}' occurred")
```

```python
connection = create_connection("sm_app.sqlite")
```

```python
create_citybikes_table = '''
CREATE TABLE IF NOT EXISTS citybikes (
    station_id TEXT PRIMARY KEY,
    station_name TEXT,
    longitude REAL,
    latitude REAL,
    Number_of_Bikes INTEGER
);
'''
```

```python
execute_query(connection, create_citybikes_table)
```

```python
create_foursquare_table = '''
CREATE TABLE IF NOT EXISTS foursquare (
    fsq_id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    category_id INTEGER,
    category_name TEXT,
    postcode INTEGER,
    distance INTEGER,
    address TEXT,
    station_id TEXT,
    FOREIGN KEY (station_id) REFERENCES citybikes (station_id)
);
'''
```

```python
execute_query(connection, create_foursquare_table)
```

```python
create_yelp_table = '''
CREATE TABLE IF NOT EXISTS yelp (
    yelp_id INTEGER PRIMARY KEY AUTOINCREMENT,
    station_id TEXT
    name TEXT,
    category_id INTEGER,
    category_name TEXT,
    review_count INTEGER,
    rating REAL,
    postcode INTEGER,
    distance INTEGER
    address TEXT
    longitude REAL,
    latitude REAL,
    price TEXT,
    FOREIGN KEY (station_id) REFERENCES foursquare (station_id)
);
'''
```

```python
execute_query(connection, create_yelp_table)
```

```python
create_yelp_fsq_table = '''
CREATE TABLE IF NOT EXISTS yelp_fsq (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    station_id TEXT
    name TEXT,
    category_id INTEGER,
    category_name TEXT,
    review_count INTEGER,
    rating REAL,
    postcode INTEGER,
    distance INTEGER
    address TEXT
    FOREIGN KEY (station_id) REFERENCES citybikes (station_id)
);
'''
```

```python
execute_query(connection, create_yelp_table)
```
