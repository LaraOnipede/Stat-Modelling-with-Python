```python
#Merge foursquare_df and yelp_df on name
# df1 = pd.merge(foursquare_df, yelp_df, on='name', how='inner')
# df1

df = pd.merge(foursquare_df, yelp_df, on='name', how='inner')
df.drop_duplicates(subset='name', inplace=True)
df
#df_1 = df1[['name']].copy()

```

```python
#use .copy() to create a copy of the merge data 
#The .fillna() method is called on the "x" columns, and the corresponding "y" columns are provided as the argument.
# This fills the missing values in the "x" columns with the values from the "y" columns where they exist. 
#This process is performed for the "address", "category_name", "postcode", and "distance" columns.
yelp_foursquare_df = df.copy()
yelp_foursquare_df['address'] = df['address_x'].fillna(df['address_y'])
yelp_foursquare_df['category_name'] = df['category_name_x'].fillna(df['category_name_y'])
yelp_foursquare_df['postcode'] = df['postcode_x'].fillna(df['postcode_y'])
yelp_foursquare_df['distance'] = df['distance_x'].fillna(df['distance_y'])
yelp_foursquare_df
```

```python
#yelp_foursquare_df.drop(['address_y', 'category_name_y', 'postcode_y', 'distance_y'], axis=1, inplace=True)
yelp_foursquare_df.drop(['address_y', 'category_name_y', 'postcode_y', 'distance_y'], axis=1, inplace=True)

yelp_foursquare_df
```

```python
yelp_foursquare_df.drop(['address_x', 'category_name_x', 'postcode_x', 'distance_x'], axis=1, inplace=True)

yelp_foursquare_df
```

```python
yelp_foursquare_df.rename(columns={
    'longitude': 'Longitude'
}, inplace=True)
yelp_foursquare_df
```
