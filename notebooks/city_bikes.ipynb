# CityBikes

 Send a request to CityBikes for the city of your choice.

```python
import requests
import json
```

```python
request_url = 'https://api.citybik.es/v2/networks'
header_dict = {'accept': 'application/JSON'}
response = requests.get(request_url, headers=header_dict)
response

#store response in a variable
citybikes = response.json()

print(citybikes)
```

```python
#print JSON response containing bike stations for Melbourne
city_id = "austin"
city_url = f'https://api.citybik.es/v2/networks/{city_id}'

response = requests.get(city_url)
austin_bikes = response.json()

print(austin_bikes)
```

```python
# Pretty-print the JSON data
print(json.dumps(citybikes, indent=4))
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
import pandas as pd
import json
```

```python
austin_bike_station = austin_bikes['network']['stations']

for station in austin_bike_station:
    latitude = station['latitude']
    longitude = station['longitude']
    num_bikes = station['free_bikes']
    
    print(f"Latitude: {latitude}, Longitude: {longitude}, Number of Bikes: {num_bikes}")
```

Put your parsed results into a DataFrame.

```python
austin_dict = {'Latitude': [], 'Longitude': [], 'Number of Bikes': []}
for station in austin_bike_station:
        latitude = station['latitude']
        longitude = station['longitude']
        num_bikes = station['free_bikes']
        
        austin_dict['Latitude'].append(latitude)
        austin_dict['Longitude'].append(longitude)
        austin_dict['Number of Bikes'].append(num_bikes)
```

```python
austinbikes_df = pd.DataFrame(austin_dict)
```
